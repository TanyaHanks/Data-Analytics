```python
print("Welcome to the Rock-Paper-Scissors game! Let's play!")

choices = ("rock", "paper", "scissors")
keep_playing = True

while keep_playing:

  player1 = input("Player 1: Enter a choice of rock, paper or scissors: ")
  while player1 not in choices:
      player1 = input("Player 1: Please enter a valid choice of rock, paper or scissors: ")

  player2 = input("Player 2: Enter a choice of rock, paper or scissors: ")
  while player2 not in choices:
      player2 = input("Player 2: Please enter a valid choice of rock, paper or scissors: ")

  if player1 == player2:
    print("It's a tie!")
  elif player1 == "paper" and player2 == "rock":
    print("Congratulations, Player 1! You are the winner!")
  elif player1 == "scissors" and player2 == "paper":
    print("Congratulations, Player 1! You are the winner!")
  elif player1 == "rock" and player2 == "scissors":
    print("Congratulations, Player 1! You are the winner!")
  else:
    print("Congratulations, Player 2! You are the winner!")


  if not input("Keep playing? y/n: ") == "y":
    keep_playing = False


print("See you soon! Thanks for playing!")



```
